```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns = ["Size", "Time"]
df_storage = pd.read_csv("Time.csv", names = columns)

x = df_storage[columns[0]]
y = df_storage[columns[1]]

plt.figure(figsize=(9, 7))
plt.plot(x, y, '-o')
plt.xlabel("Size of dataset", fontsize=15)
plt.ylabel("Time complexity (s)", fontsize=15)
```

```python
columns = ["Size", "RAM"]
df_storage = pd.read_csv("RAM.csv", names = columns)    # RAM usage of fullfilled array 'numpy.ndarray.nbytes'

x = df_storage[columns[0]]
y = df_storage[columns[1]]

plt.figure(figsize=(9, 7))
plt.plot(x, y, '-o')
plt.xlabel("Size of dataset", fontsize=15)
plt.ylabel("Memory complexity (GB)", fontsize=15)
```
